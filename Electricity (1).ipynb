```python
def custom_plot(kind = None ,x = None, y = None, label = None, ax=None, **plt_kwargs):
    if ax is None:
        ax = plt.gca()
    if kind == None:
        ax.xaxis.set(
            major_locator=mdates.DayLocator(),
            major_formatter=mdates.DateFormatter("\n\n%A"),
            minor_formatter=mdates.DateFormatter("%H"),

        )
        ax.plot(x, y, color = 'black')
    else:
        ax.xaxis.set(
            major_locator=mdates.DayLocator(),
            major_formatter=mdates.DateFormatter("\n\n%A"),
            minor_locator=mdates.HourLocator((3,6, 9 ,12, 15 ,18,21)),
            minor_formatter=mdates.DateFormatter("%H"),
        )
        ax.plot(electric_df['Time'], electric_df['Sub_metering_1'], label = 'Sub_metering_1', color = 'black')
        ax.plot(electric_df['Time'], electric_df['Sub_metering_2'], label = 'Sub_metering_2', color = 'red')
        ax.plot(electric_df['Time'], electric_df['Sub_metering_3'], label = 'Sub_metering_3', color = 'blue')
        
        ax.legend(loc="upper right",fontsize = 11)
        
    ax.set_ylabel(label, fontsize = 12)
    
    return(ax)
```

```python
fig, ax2 = plt.subplots(figsize = (25,16))
def plot2(ax):
    custom_plot(1,electric_df['Time'], electric_df['Global_active_power'],'Global Active Power(kilowatts)',ax)
plot2(ax2)
```

```python
fig, ax2 = plt.subplots(figsize = (25,16))
def plot3(ax):
    custom_plot(2,ax)
plot3(ax2)
```

```python
def plot3():
    fig, ax3 = plt.subplots(figsize = (25,16))
    ax3.plot(electric_df['Time'], electric_df['Sub_metering_1'], label = 'Sub_metering_1', color = 'black')
    ax3.plot(electric_df['Time'], electric_df['Sub_metering_2'], label = 'Sub_metering_2', color = 'red')
    ax3.plot(electric_df['Time'], electric_df['Sub_metering_3'], label = 'Sub_metering_3', color = 'blue')
    
    ax3.xaxis.set(
        major_locator=mdates.DayLocator(),
        major_formatter=mdates.DateFormatter("\n\n%A"),
        minor_locator=mdates.HourLocator((3,6, 9 ,12, 15 ,18,21)),
        minor_formatter=mdates.DateFormatter("%H"),
    )
    plt.ylabel('Energy sub metering', fontsize = 20)
    ax3.legend(loc="upper right",fontsize = 20)
    plt.show()
```
